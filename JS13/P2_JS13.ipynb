# Praktikum 2

Pada praktikum ini menggunakan library Keras untuk menggunakan JST. Keras adalah API tingkat tinggi untuk membangun JST dengan mudah, sedangkan TensorFlow adalah framework yang mendukung Keras.

## Langkah:

1. Import library.
2. Load dataset.
3. Bangun model.
4. Kompilasi dan latih model.
5. Evaluasi hasil.


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model.fit(X_train, y_train, epochs=50, batch_size=8)

# Evaluasi
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi: {acc}")
```

### **Hasil Praktikum Awal (Konfigurasi Default)**

Model awal menggunakan:
- Hidden layer 1: 10 neuron (ReLU)
- Hidden layer 2: 8 neuron (ReLU)

**Akurasi awal:**  
`0.7667`

# Tugas 2


- Ubah jumlah neuron hidden layer.
- Bandingkan akurasi dengan konfigurasi awal.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model (DIUBAH jumlah neuron)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(4,)),  # dari 10 → 16
    tf.keras.layers.Dense(12, activation='relu'),                    # dari 8 → 12
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model.fit(X_train, y_train, epochs=50, batch_size=8, verbose=1)

# Evaluasi
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi (Tugas 2): {acc}")
```

### **Konfigurasi**

Hidden layer diubah menjadi:
- Hidden layer 1: **16 neuron (ReLU)**
- Hidden layer 2: **12 neuron (ReLU)**

### **Hasil**
- **Akurasi:** `0.9667`



### **Analisis**
| Model | Hidden Layer | Aktivasi | Akurasi |
|-------|--------------|----------|---------|
| Praktikum awal | 10 → 8 | ReLU | **0.7667** |
| Tugas 2 | 16 → 12 | ReLU | **0.9667** |

### **Kesimpulan**

- Dengan menambah jumlah neuron pada kedua hidden layer, akurasi meningkat signifikan dari **76.67% → 96.67%**.  
- Penambahan neuron membuat jaringan memiliki kapasitas representasi lebih besar sehingga lebih mampu mempelajari pola pada dataset Iris.


# Tugas 3

- Bandingkan Sigmoid vs ReLU pada dataset Iris.
- Catat perbedaan loss dan akurasi.

## Hidden layer Sigmoid

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model (SIGMOID)
model_sigmoid = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='sigmoid', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model_sigmoid.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model_sigmoid.fit(X_train, y_train, epochs=50, batch_size=8, verbose=1)

# Evaluasi
loss_s, acc_s = model_sigmoid.evaluate(X_test, y_test)
print(f"[SIGMOID] Loss: {loss_s}, Akurasi: {acc_s}")
```

## Hidden layer ReLU

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model (RELU)
model_relu = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model_relu.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model_relu.fit(X_train, y_train, epochs=50, batch_size=8, verbose=1)

# Evaluasi
loss_r, acc_r = model_relu.evaluate(X_test, y_test)
print(f"[RELU] Loss: {loss_r}, Akurasi: {acc_r}")
```

Pada tugas 3 dilakukan pengujian dua model:
- Model A → Hidden layer menggunakan **Sigmoid**
- Model B → Hidden layer menggunakan **ReLU**

---

### **Model A – Sigmoid**

**Hasil:**
- Loss: `0.6434`
- Akurasi: `0.6667`

### **Model B – ReLU**

**Hasil:**
- Loss: `0.5053`
- Akurasi: `0.9000`

---

### **Perbandingan Sigmoid vs ReLU**

| Model | Hidden Activation | Loss | Akurasi |
|--------|-------------------|--------|-----------|
| Sigmoid | Sigmoid–Sigmoid | **0.6434** | **0.6667** |
| ReLU | ReLU–ReLU | **0.5053** | **0.9000** |

---

### **Kesimpulan**

Aktivasi ReLU memberikan performa yang jauh lebih baik dibanding Sigmoid pada dataset Iris. ReLU menghasilkan **loss lebih rendah** dan **akurasi lebih tinggi (90%)**, sedangkan Sigmoid hanya mencapai akurasi sekitar 66%. ReLU lebih mampu menghindari masalah vanishing gradient sehingga proses training lebih efektif.
